### 1. 유클리드 거리(Euclidean distance)

유클리드 거리는 문서의 유사도를 구할 때 자카드 유사도나 코사인 유사도만큼 유용한 방법은 아닙니다.

다차원 공간에서 두개의 점 p와q 가 각각 p = (p1,p2,p3,...pn)과
q = (q1,q2,q3,...qn)의 좌표를 가질 때 두 점 사이의 거리를 계산하는 유클리드 거리 공식은 다음과 같습니다.

```python
from IPython.display import Image
```

```python
Image("euclidean.PNG")
```

2차원 좌표 평면 상에서 두 점 와 사이의 직선 거리를 구하는 문제입니다.<br>
피타고라스의 정리를 통해 두 점 사이의 거리를 구하는 것과 동일합니다.<br>
<b>다시 원점으로 돌아가서 여러 문서에 대해서 유사도를 구하고자 유클리드 거리 공식을 사용한다는 것은, 앞서 본 2차원을 단어의 총 개수만큼의 차원으로 확장하는 것과 같습니다. 예를 들어 아래와 같은 DTM이 있다고 합시다.</b>

```python
Image("io.PNG")
```

```python
import numpy as np
def dist(x,y):   
    return np.sqrt(np.sum((x-y)**2))

doc1 = np.array((2,3,0,1))
doc2 = np.array((1,2,3,1))
doc3 = np.array((2,1,2,2))
docQ = np.array((1,1,0,1))

print(dist(doc1,docQ))
print(dist(doc2,docQ))
print(dist(doc3,docQ))
```

## 2. 자카드 유사도(jaccard similarity)



```python
Image("oi.PNG")
```

```python
# 다음과 같은 두 개의 문서가 있습니다.
# 두 문서 모두에서 등장한 단어는 apple과 banana 2개.
doc1 = "apple banana everyone like likey watch card holder"
doc2 = "apple banana coupon passport love you"

# 토큰화를 수행합니다.
tokenized_doc1 = doc1.split()
tokenized_doc2 = doc2.split()

# 토큰화 결과 출력
print(tokenized_doc1)
print(tokenized_doc2)
```

```python
union = set(tokenized_doc1).union(set(tokenized_doc2))
print(union)
```
