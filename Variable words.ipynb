<h3>1. 단어의 표현 방법</h3>

<p>
    표현 방법은 크게 국소 표현 방법과 분산 표현 방법으로 나뉩니다.<br><br>
    <b>국소 표현 방법 : 해당 단어 그 자체만 보고, 특정값을 맵핑하여 단어를 표현하는 방법<br>
        분산 표현 방법 : 그 단어를 표현하고자 주변을 참고하여 단어를 표현하는 방법
</p>

<b>예를 들어 강아지, 귀여운, 사랑스러운 단어에서 각 단어에 1번, 2번, 3번 등과 같은 숫자를 맵핑하는 방법이 국소 표현 방법<br>
    강아지라는 단어 근처에는 주로 귀여운, 사랑스러운이라는 단어가 자주 등장하므로 강아지는 귀엽다 라는 느낌으로 정의하는것이 분산 표현 방법
