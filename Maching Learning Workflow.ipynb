<h3>1) 수집(Acquisition)</h3>

<p>머신 러닝을 위해서 기계에 학습시켜야 할 데이터를 <b>말뭉치 또는 코퍼스</b>라고 부른다.</p>

<h3>2) 점검 및 탐색(Inspection and exploration)</h3>

<p>
    데이터가 수집되었다면, 데이터를 점검하고 탐색하는 단계입니다.<br>
    이 단계를 <b>탐색적 데이터 분석 단계</b>라고도 한다.<br>
    데이터의 특징과 내재하는 구조적 관계를 알아내는 과정을 의미합니다. 시각화, 간단한 통계 테스트를 진행하기도 한다.
</p>

<h3>3) 전처리 및 정제(Preprocessing and Cleaning)</h3>

<p>이 단계는 많은 단계를 포함하고 있는데, 가령 자연어 처리라면<br>
    토큰화, 정제, 정규화, 불용어 제거 등의 단계를 포함합니다.</p>

<h3>4) 모델링 및 훈련(Modeling and Training)</h3>

<p>
    머신 러닝에 대한 코드를 작성하는 단계인 모델링 단계에 들어갑니다.<br>
    모든 데이터를 기계에 학습시키지 말고 일부는 테스트용으로 남겨두고 훈련용 데이터만 훈련에 사용해야 합니다.<br>
</p>

```python
from IPython.display import Image
```

```python
Image("Modeling.PNG")
```

<p>비유하자면 훈련용은 <b>학습지</b>,검증용은 <b>모의고사</b>,테스트용은 <b>수능 시험</b>이라고 볼 수 있습니다.<br>
    검증용 데이터가 얼마나 제대로 학습이 되었는지를 판단하는 용으로 사용되며 <br>
    테스트용 데이터는 모델의 최종 성능을 평가하는 데이터로 검증용 데이터는 모델의 성능을 개선하는데 사용하지만,<br>
    테스트용 데이터는 모델의 성능을 개선하는 일에 사용되는 것이 아닙니다.<br>
    쉽게 말해 시험에 비유하자면 이제는 채점하는 단계입니다.

<h3>5) 평가(Evaluation)</h3>

<p>기계가 다 학습이 되었다면 테스트용 데이터로 성능을 평가하게 됩니다.<br>
    평가 방법은 기계가 예측한 데이터가 테스트용 데이터의 실제 정답과 얼마나 가까운지를<br> 측정합니다.</p>

<h3>6) 배포(Deployment)</h3>

<p>평가 단계에서 기계가 성공적으로 훈련이 된 것으로 판단된다면, 완성된 모델이 배포되는 단계이다.<br>
    실제로는 거의 모든 단계에서 전 단계로 돌아가는 상황이 발생할 수 있습니다.
</p>
